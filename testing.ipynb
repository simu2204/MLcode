```python
df['director']=df['director'].fillna('director not specified')
df.head()
```

```python
directors_list = pd.DataFrame()
print(directors_list)
```

```python
directors_list = df['director'].str.split(',', expand=True).stack()
print(directors_list)
```

```python
directors_list = directors_list.to_frame()
directors_list.columns = ['directors list']
directors_list.head()
```

```python
directors_list.columns = ['Director']
print(directors_list)
```

```python
directors = directors_list.groupby(['Director']).size().reset_index(name='Total Count')
print(directors)
```

```python
directors = directors[directors.Director != 'Director not specified']
```

```python
print(directors)
```

```python
directors = directors.sort_values(by=['Total Count'], ascending = False)
print(directors)
```

```python
top5Directors = directors.head()
print(top5Directors)
```

```python
import matplotlib.pyplot as plt

top5Directors = top5Directors.sort_values(by=['Total Count'])

plt.figure(figsize=(5,5))
plt.barh(top5Directors['Director'], top5Directors['Total Count'], color='skyblue')

plt.xlabel('Total Count')
plt.ylabel('Director')
plt.title('Top 5 Directors on Netflix')

plt.show()
```

```python

df['cast'] = df['cast'].fillna('No cast specified')

cast_df = df['cast'].str.split(',', expand=True).stack()
cast_df = cast_df.to_frame()
cast_df.columns = ['Actor']

actors = cast_df.groupby(['Actor']).size().reset_index(name='Total Count')


actors = actors[actors.Actor != 'No cast specified']

actors = actors.sort_values(by=['Total Count'], ascending=False)

top5Actors = actors.head()

top5Actors = top5Actors.sort_values(by=['Total Count'])

plt.figure(figsize=(5, 6))
plt.barh(top5Actors['Actor'], top5Actors['Total Count'], color='skyblue')

plt.xlabel('Total Count')
plt.ylabel('Actor')
plt.title('Top 5 Actors on Netflix')


plt.show()
```

```python
df1 = df[['type', 'release_year']]
df1 = df1.rename(columns = {"release_year":"Release Year", "type": "Type"})
df2 = df1.groupby(['Release Year', 'Type']).size().reset_index(name='Total Count')
```

```python
print(df2)
```

```python
import matplotlib.pyplot as plt

# Filter the data to only include content from 2000 onwards
df2 = df2[df2['Release Year'] >= 2000]

# Plot the trend of content produced on Netflix every year
plt.figure(figsize=(5,5))  # Set the figure size

# Plot each 'Type' separately with different colors
for content_type in df2['Type'].unique():
    filtered_df = df2[df2['Type'] == content_type]
    plt.plot(filtered_df['Release Year'], filtered_df['Total Count'], label=content_type)

# Add labels and title
plt.xlabel('Release Year')
plt.ylabel('Total Count')
plt.title('Trend of Content Produced on Netflix Every Year')

# Display the legend
plt.legend(title='Type')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from textblob import TextBlob

# Your DataFrame 'df' containing release year and description

df3 = df[['release_year', 'description']]
df3 = df3.rename(columns = {'release_year': 'Release Year', 'description': 'Description'})

# Sentiment analysis using TextBlob
for index, row in df3.iterrows():
    d = row['Description']
    testimonial = TextBlob(d)
    p = testimonial.sentiment.polarity
    if p == 0:
        sent = 'Neutral'
    elif p > 0:
        sent = 'Positive'
    else:
        sent = 'Negative'
    df3.loc[index, 'Sentiment'] = sent

# Group by Release Year and Sentiment
df3 = df3.groupby(['Release Year', 'Sentiment']).size().reset_index(name='Total Count')

# Filter for release years greater than 2005
df3 = df3[df3['Release Year'] > 2005]

# Create the bar plot using Matplotlib
plt.figure(figsize=(6, 6))  # Set the figure size

# Plot the bars for each sentiment category
colors = {'Positive': 'green', 'Neutral': 'gray', 'Negative': 'red'}

for sentiment in df3['Sentiment'].unique():
    sentiment_data = df3[df3['Sentiment'] == sentiment]
    plt.bar(sentiment_data['Release Year'], sentiment_data['Total Count'],
            label=sentiment, color=colors[sentiment])

# Add labels and title
plt.xlabel('Release Year')
plt.ylabel('Total Count')
plt.title('Sentiment Analysis of Content on Netflix')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')

# Display the plot
plt.tight_layout()  # Adjust layout to make sure everything fits
plt.show()
```
